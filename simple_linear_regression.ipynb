```python
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#equation
#y=mx+c
m=50
c=36
```

```python
x=np.random.randint(0,50000,200) # generating random x_values
y=np.array(list(map(lambda x:m*x+c,x))) # applying the equation on x to get y values
y
```

```python
#adding noice on y
mean=0
std=20 #small value for accurate model
noice=np.random.normal(mean,std,len(y))
noiced_y=y+noice
####the best noice type,mean,std??
```

```python
#split data 80:20
x=x.reshape(-1,1)
#########whhyyyyy??????????
X_train, X_test, y_train, y_test = train_test_split(
    x, noiced_y, test_size=0.2, random_state=0)
```

```python
#model fit
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
print(model.coef_,model.intercept_)
```

```python
predictions=model.predict(X_test)
```

```python
sns.scatterplot(x=X_test.ravel(),y=y_test)
sns.lineplot(x=X_test.ravel(),y=predictions,color='red')
####why x_test is 2d??????????
```
